```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as plt
data= pd.read_csv("/content/black_white_wage_gap.csv")

#Distribution plot
sns.displot(data['white_median'],bins=10,kde=True,rug=False)
#Box plot
data["black"]=pd.cut(x=data['black_average'],bins=[0,3,17,63,99],labels=["Baby","Child","Adult","Elderly"])
sns.catplot(x="white_men_median",y="year",kind="box",data=data)
#violin plot
sns.violinplot(x="black_women_average",y="white_average",data=data,size=5)
#line plot
test_data=pd.DataFrame(dict(time=pd.date_range("2017-1-1",periods=500),value=np.random.randn(500).cumsum()))
sns.relplot(x="time",y="value",kind="line",data=test_data,height=5, aspect=15/6)
#bar plot
sns.catplot(x="white_average",y="black_average",kind="bar",data=data)
#scatter plot
sns.jointplot(x="year",y="white_men_median",kind="scatter",data=data)
# pair plot
sns.pairplot(data)
#creating dataframe, showing their correlations
data1 = {'A': np.random.rand(100), 'B': np.random.rand(100)}
dataframe1 = pd.DataFrame(data1)
data2 = {'C': np.random.rand(100), 'D': np.random.rand(100)}
dataframe2 = pd.DataFrame(data2)
correlation1 = dataframe1.corr(method='pearson')
correlation2 = dataframe2.corr(method='pearson')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(correlation1, annot=True, cmap='coolwarm', fmt=".2f", ax=ax1)
ax1.set_title("Correlation Heatmap - DataFrame 1")
sns.heatmap(correlation2, annot=True, cmap='coolwarm', fmt=".2f", ax=ax2)
ax2.set_title("Correlation Heatmap - DataFrame 2")

plt.tight_layout()
plt.show()
```
