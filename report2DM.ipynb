```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import cross_val_score, KFold
import matplotlib.pyplot as plt

data = load_iris()
X, y = data.data, data.target

unpruned_tree = DecisionTreeClassifier(random_state=42)
pruned_tree = DecisionTreeClassifier(random_state=42, ccp_alpha=0.015)

kf = KFold(n_splits=15, shuffle=True, random_state=42)

unpruned_tree.fit(X, y)
pruned_tree.fit(X, y)

unpruned_scores = cross_val_score(unpruned_tree, X, y, cv=kf)
pruned_scores = cross_val_score(pruned_tree, X, y, cv=kf)

plt.figure(figsize=(12, 6))
plot_tree(unpruned_tree, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.title("Unpruned Decision Tree")
plt.show()

plt.figure(figsize=(12, 6))
plot_tree(pruned_tree, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.title("Pruned Decision Tree")
plt.show()

print("Unpruned Tree Accuracy:", unpruned_scores.mean())
print("Pruned Tree Accuracy:", pruned_scores.mean())
```
